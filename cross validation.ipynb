0) Очищаем данные X и y (при необходимости)
1) Разбиваем X и y на обучающий и тестовый наборы
2) Обучаем объект Scaler на обучающих данных X
3) Применяем масштабирование (scale) на тестовых данных X
4) Создаем модель
5) Обучаем модель на обучающих данных Х
6) Оцениваем модель на тестовых данных Х (создавая предсказания и сравнивая с y_test)

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=101
)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=100)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE = mean_squared_error(y_test, y_pred)
```

```python
MSE
```

есть ли более удачные значения alpha?

```python
model_two = Ridge(alpha=1)
```

```python
model_two.fit(X_train, y_train)
```

```python
y_pred_two = model_two.predict(X_test)
```

```python
MSE = mean_squared_error(y_test, y_pred_two)
```

```python
MSE
```

### **Кросс-валидация - cross_val_score**

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=101
)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
model = Ridge(alpha=100)
```

А дальше шаги уже по-другому!

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(estimator=model, X=X_train,y=y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
scores
```

```python
abs(scores.mean())
```

```python
# alpha is different
model = Ridge(alpha=1)
```

```python
scores = cross_val_score(estimator=model, X=X_train,y=y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
abs(scores.mean())
```

```python
model.fit(X_train, y_train)
```

```python
y_final_test_pred = model.predict(X_test)
```

```python
MSE = mean_squared_error(y_test, y_final_test_pred)
```

```python
MSE
```
