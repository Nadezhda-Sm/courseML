# **Unsupervised Learning - Обучение БЕЗ учителя**

При помощи **неразмеченных (unlabeled)** данных найти закономерности в данных, собирать кластеры или существенные компоненты

1) **Кластеризация**
   
   С помощью признаков сгруппировать строки в отделньые кластеры
2) **Уменьшение размерности**
   
   Уменьшить размерность пространства признаков

Метрики из обучения с учителем НЕ ПОДХОДЯТ. Ведь у нас нет истинных значений, с которыми можно было бы сравнивать

***ЦЕЛЬ*** Найти закономерности в данных!

## **Этапы машинного обучения без учителя**

Реальный мир --> Сырые данные --> Сбор и хранение данных (данные не будут размечены (labeled)) --> Очистка и организация данных

--> Исследование данных --> 

1) Кластеризация

2) Уменьшение размерности

*Кластеризация:* 

если есть неразмеченные данные, то можем ли мы их собрать похожие точки данных вместе, чтобы "найти" возможные значени целево переменной для кластеров?

*Уменьшение размерности:* 

если есть неразмеченные данные, то можем ли мы уменьшить количество признаков, собирая их в новые компоненты? Могут ли эти новые компоненты дать нам новые знания о данных?
