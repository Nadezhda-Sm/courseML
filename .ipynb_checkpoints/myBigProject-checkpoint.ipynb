## Функция вычисления retention

```python
def calculate_retention(reg_data, auth_data, cohort_period='D', observation_days=30):          
    
    # Создание когорт
    if cohort_period == 'D':
        reg_data['cohort'] = reg_data['reg_ts'].dt.date
    elif cohort_period == 'W':
        reg_data['cohort'] = reg_data['reg_ts'].dt.to_period('W').dt.start_time.dt.date
    elif cohort_period == 'M':
        reg_data['cohort'] = reg_data['reg_ts'].dt.to_period('M').dt.start_time.dt.date
    
    # Создание датафрейма с когортами пользователей
    cohorts = reg_data[['uid', 'cohort']].copy()
    
    # Добавление информации о визитах
    user_activity = pd.merge(auth_data, cohorts, on='uid', how='inner')
    user_activity['activity_date'] = user_activity['auth_ts'].dt.date
    
    # Расчет дня от регистрации
    user_activity['days_since_reg'] = (user_activity['activity_date'] - user_activity['cohort']).dt.days
    
    # Удаляем отрицательные значения (если они есть из-за ошибок в данных)
    user_activity = user_activity[user_activity['days_since_reg'] >= 0]
    
    # Группировка по когортам и дням
    cohort_pivot = user_activity.groupby(['cohort', 'days_since_reg'])['uid'].nunique().reset_index()
    
    # Создание pivot таблицы для retention
    retention_table = cohort_pivot.pivot_table(
        index='cohort', 
        columns='days_since_reg', 
        values='uid'
    )
    
    # Рассчитываем размеры когорт (первый столбец сводной таблицы)
    cohort_sizes = retention_table.iloc[:, 0] 
    # Вычисляем коэффициенты удержания, деля на размер когорты
    retention_rate = retention_table.div(cohort_sizes, axis=0)
    
    # Ограничение по количеству дней наблюдения
    retention_rate = retention_rate.loc[:, :observation_days]
    
    return retention_rate
```

## Функция для визуализации retention

```python
# Функция для визуализации retention
def plot_retention_heatmap(retention_data):
    plt.figure(figsize=(16, 10))
    sns.heatmap(retention_data, 
                annot=True, 
                fmt='.1f', 
                cmap='RdYlGn',
                cbar_kws={'label': 'Retention Rate (%)'})
    plt.title('Retention Rate by Cohort and Day', fontsize=16, pad=20)
    plt.xlabel('Days Since Registration')
    plt.ylabel('Cohort')
    plt.tight_layout()
    plt.show()
```

### Анализ данных

```python
df.head()
```

```python
shape = df.shape
n_unique_users = df['user_id'].nunique()
group_counts = df['testgroup'].value_counts()

overall_revenue_sum = df['revenue'].sum()
revenue_desc = df['revenue'].describe()
```

```python
# Определение платящих
df['is_payer'] = df['revenue'] > 0
```

```python
# Агреггируем данные
grp = df.groupby('testgroup').agg(
    users=('user_id', 'count'),
    total_revenue=('revenue', 'sum'),
    arpu=('revenue', 'mean'), 
    median_revenue=('revenue', 'median'),
    std_revenue=('revenue', 'std'),
    payers=('is_payer', 'sum')
)
# CR - Conversion Rate (доля платящих пользователей)
grp['cr'] = grp['payers'] / grp['users']
# ARPPU - Average Revenue Per Paying User (средняя выручка на платящего пользователя)
grp['arppu'] = grp.apply(lambda row: row['total_revenue'] / row['payers'] if row['payers'] > 0 else 0, axis=1)
```

```python
grp['arppu_payers'] = grp['total_revenue'] / grp['payers'].replace(0, np.nan)
```

#### Статистические критерии

```python
rev_a = df.loc[df['testgroup'] == 'a', 'revenue']
rev_b = df.loc[df['testgroup'] == 'b', 'revenue']
```

##### t-test для ARPU (revenue per user)

```python
t_stat_arpu, t_p_arpu = stats.ttest_ind(rev_a, rev_b, equal_var=False)
```

##### z-test для CR (payer rate)

```python
count = np.array([grp.loc['a', 'payers'], grp.loc['b', 'payers']])
nobs = np.array([grp.loc['a', 'users'], grp.loc['b', 'users']])
p1 = count[0] / nobs[0]
p2 = count[1] / nobs[1]
p_pool = count.sum() / nobs.sum()
se = np.sqrt(p_pool * (1 - p_pool) * (1 / nobs[0] + 1 / nobs[1]))
z_cr = (p1 - p2) / se
p_z_cr = 2 * (1 - norm.cdf(abs(z_cr)))
```

##### Статистический анализ для ARPPU (только платящие пользователи)

```python
payers_revenue_a = df[(df['testgroup'] == 'a') & (df['is_payer'])]['revenue']
payers_revenue_b = df[(df['testgroup'] == 'b') & (df['is_payer'])]['revenue']

# Проверяем, что в обеих группах есть платящие пользователи для теста
if not payers_revenue_a.empty and not payers_revenue_b.empty:
    t_stat_arppu, t_p_arppu = stats.ttest_ind(payers_revenue_a, payers_revenue_b, equal_var=False)
else:
    t_stat_arppu, t_p_arppu = np.nan, np.nan
    print("В одной из групп нет платящих пользователей, тест ARPPU не может быть выполнен.")
```

#### Результаты

```python
print('shape:', df.shape)
print('unique users:', df['user_id'].nunique())
print('\nGroup counts:')
print(df['testgroup'].value_counts())
print('\nOverall revenue sum:', df['revenue'].sum())
print('\nRevenue description:')
print(df['revenue'].describe())
print('\nGrouped metrics:')
print(grp)
print('\n--- Статистические тесты ---')
print(f'ARPU (revenue per user) A: {grp.loc["a", "arpu"]:.4f}, B: {grp.loc["b", "arpu"]:.4f}')
print('T-test for ARPU: t=%.4f, p=%.6f' % (t_stat_arpu, t_p_arpu))
print(f'\nCR (payer rate) A: {grp.loc["a", "cr"]:.6f}, B: {grp.loc["b", "cr"]:.6f}')
print('Z-test for CR: z=%.4f, p(two-sided)=%.6f' % (z_cr, p_z_cr))

print(f'\nARPPU (payers only) A: {grp.loc["a", "arppu"]:.4f}, B: {grp.loc["b", "arppu"]:.4f}')
print('T-test for ARPPU: stat=%.4f, p=%.6f' % (t_stat_arppu, t_p_arppu))
```

На основе результатов A/B теста, мы не обнаружили статистически значимого увеличения общей выручки на пользователя (ARPU) ни в одной из групп. Это означает, что хотя в тестовой группе B наблюдался прирост ARPU на 5%, мы не можем с уверенностью утверждать, что это не случайность.

  - Контрольная группа A: Показала более высокий коэффициент конверсии (CR) в платящих пользователей.

Таким образом, выбор лучшего набора предложений зависит от вашей основной стратегической цели:
  - Если приоритет — увеличение числа платящих пользователей: Набор предложений A является лучшим выбором.
  - Если приоритет — стимулирование платящих пользователей тратить больше: Набор предложений B может быть предпочтительнее, хотя он и приводит к меньшему числу платящих пользователей.

Поскольку общая выручка на пользователя (ARPU) не показала статистически значимой разницы, оба набора имеют свои преимущества, компенсирующие недостатки друг друга. Возможно, потребуется провести дополнительные исследования или сфокусироваться на одной из этих целей для дальнейшего тестирования.


## 3) Метрики для оценки тематического события

1. Активность игроков:

  •  Daily Active Users (DAU): Количество уникальных пользователей, заходивших в игру каждый день во время события. Покажет, насколько событие привлекало игроков каждый день.  
  •  Monthly Active Users (MAU): Количество уникальных пользователей, заходивших в игру хотя бы раз в месяц во время события. Показывает общий охват события.  
  •  Peak Concurrent Users (PCU): Максимальное количество игроков, одновременно находящихся в игре во время события. Показывает максимальную нагрузку и популярность события в определенный момент.  
  •  Session Length: Средняя продолжительность игровой сессии. Увеличение session length может свидетельствовать о том, что событие увлекает игроков.  
  •  Sessions per User: Среднее количество игровых сессий на одного пользователя. Показывает, как часто игроки возвращались в игру во время события.
  
  
2. Вовлеченность игроков:

  •  Event Participation Rate: Процент игроков, принявших участие в событии (например, выполнивших квест, сыгравших в специальный режим).  
  •  Average Event Progress: Средний прогресс игроков в рамках события (например, сколько уровней они прошли, сколько заданий выполнили). Показывает, насколько игроки вовлечены в выполнение целей события.  
  •  Event Completion Rate: Процент игроков, полностью завершивших событие. Показывает, насколько достижимы цели события.  
  •  Social Interactions: Количество взаимодействий между игроками (например, отправка подарков, участие в клановых мероприятиях, общение в чате). Увеличение социальных взаимодействий может свидетельствовать о том, что событие способствует формированию сообщества.  
  •  New Content Usage: Процент игроков, использовавших новый контент, добавленный в рамках события (например, новых персонажей, оружие, локации). Показывает, насколько интересен новый контент игрокам.
  
  
3. Монетизация:

  •  Event Revenue: Общий доход, полученный во время события (например, от продажи внутриигровых предметов, подписок). Показывает, насколько событие способствовало увеличению дохода.
  •  Paying Conversion Rate: Процент игроков, совершивших покупку во время события. Показывает, насколько событие стимулировало игроков тратить деньги.  
  •  Average Revenue Per Paying User (ARPPU): Средний доход с каждого пользователя, совершившего покупку во время события.  
  •  Average Revenue Per Daily Active User (ARPDAU): Средний доход на одного активного пользователя в день.  
  •  Average Revenue Per User (ARPU): Средний доход на одного пользователя за все время события.
  
  
4. Удержание:

  •  Retention Rate (Event Participants): Процент игроков, принявших участие в событии, которые вернулись в игру через 7, 14, 30 дней после окончания события. Показывает, насколько событие повлияло на долгосрочное удержание игроков.
  
  
  
5. Пользовательские отзывы:

  •  Sentiment Analysis: Анализ тональности отзывов игроков о событии (например, положительные, отрицательные, нейтральные). Позволяет понять, что понравилось и что не понравилось игрокам в событии.  
  •  Support Ticket Volume: Количество обращений в службу поддержки, связанных с событием. Увеличение количества обращений может свидетельствовать о проблемах в событии.
