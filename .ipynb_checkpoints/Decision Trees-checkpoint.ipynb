# **Деревья решений (Decision Trees)**

 - Деревья решений (Decision Trees)
 - Случайные леса (Random Forests)
 - Расширяемые деревья (Boosted Trees) -- градиентный спуск 

*Обучение деревьев решений (Decision Tree Learning)* - статистическое моделирование с применением деревьев, где решения в каждом узле применяются некоторого условия (некоторой метрики)

**CART (Classification and Regression Trees)** - наборы правил для построения деревьев решений. Используются и в sklearn!

- Кросс-валидация деревьев
- Усечение деревьев (pruning)
- Суррогатные разбиения
- Оценки важности переменных
- Поиск линейных разделений

*Терминология Деревьев решений*

-Разбиение (splitting) - в каждом узле дерева принимаем нек.решение; некторое условие возвращает True/False
-Узлы (Nodes) - прямоугольнички
- Корневой узел (Root Node)
- Листовые / Конечные узлы (Leaf/ Terminal nodes) - узлы, где мы возвращаем значение целевой переменной, не делаем разбиение, принимаем решение
- Родительские/ Дочерние узлы (Parent/Childern Nodes) - из Род. узла выходят Доч. узлы
- Ветви / Поддеревья (Tree burnches / Subtrees)
- Усечение (Pruning) - заменить разбиение на листовой узел --> избежать overfitting

Как при построении деревьев выбираются критерии разбиения? 

**Gini Impurity (загрязнение Джини, индекс Джини, примесь Джини)** - характ-т то, насколько информация в наборе данных является чистой. 

С точки зрения классификации, мы можем воспринимать это как однородность классов. 

Определение из Интернета:

Gini Impurity - мера «неоднородности» или «примеси» данных в узлах решающих деревьев, используемая для выбора наилучшего расщепления (split). Он показывает вероятность неправильной классификации элемента, если он будет случайно выбран из узла. Чем ниже значение индекса, тем «чище» узел и лучше разделение. 

$$
Gini(Q) = 1 - \sum_{c \in C} p_i*(1-p_i)^2
$$
,где:
- Q - набор данных
- С - набор классов

$$
p_c = 1/N_Q * \sum_{x \in Q} (y_{class} = c)
$$

Gini = 0 - данные полностью однородные, все данные принадлежат к одному классу

Цель: минимизировать Gini Impurity на листовых узлах

1) Вычисляем Gini для каждого листового узла
2) Умножаем каждое полученное значение на средневзвешенное
3) G(Q) = сумма зн-ий из п.2)

Threshhold - некое пороговое значение G(Q), и мы будем выполнять разбиение только в том случае, если оно даст G(Q) <= этому пороговому значению.

Это и есть усечение (Pruning), борьба с overfitting .

```python
X = pd.get_dummies(df.drop('species', axis=1), drop_first=True)
```

```python
X.head()
```

```python
y = df['species']
```

Масштабирование не нужно, т.к. выбор будет идти только по одному признаку, в каждом узле участвует только один признак.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
base_preds = model.predict(X_test)
```

```python
base_preds
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

```python
print(classification_report(y_test, base_preds))
```

```python
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
```

```python
model.feature_importances_
```

```python
pd.DataFrame(index=X.columns, data=model.feature_importances_, columns=['Важность признаков']).sort_values(by='Важность признаков', ascending=False)
```

```python
from sklearn.tree import plot_tree
```

X[2] - flipper_length_mm

samples - сколько точек пришло на вход в узел, в корневой узел попадают все данные

value - сколько точек принадлежит к каждому из классов целевой переменной. сумма значений из value = samples

```python
plt.figure(figsize=(12,8), dpi=200)
plot_tree(model, feature_names=X.columns, filled=True);
```

```python
X_train.columns
```

Поиграемся с гиперпараметрами

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test, model_preds))
    print()
    plt.figure(figsize=(12,8), dpi=200)
    plot_tree(model, feature_names=X.columns, filled=True);
```

```python
report_model(model)
```

глубина дерева

```python
pruned_tree = DecisionTreeClassifier(max_depth=3) #верхний лимит кол-ва уровней, не об-но его достигать
```

```python
pruned_tree.fit(X_train, y_train)
```

```python
report_model(pruned_tree)
```

```python
max_leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
max_leaf_tree.fit(X_train, y_train)
```

```python
report_model(max_leaf_tree)
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
```

```python
entropy_tree.fit(X_train, y_train)
```

```python
report_model(entropy_tree)
```

```python
ConfusionMatrixDisplay.from_estimator(entropy_tree, X_test, y_test)
```
