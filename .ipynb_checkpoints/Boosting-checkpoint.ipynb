# **Boosting**

Decision Tree --> Random Forest for perfomance improvement



**Boosting** is another technique for Decision Trees' perfomance improvement

Boosting is an *meta-algorythm* and it can be used for many ML models

Бустинг это не алгоритм машинного обучения, а методология для применения к существующим алгоритмам МО, чаще всего к деревьям решений, т.к. для них получаются наилучшие результаты.

- Бустинг и мета-обучение
- Теория AdaBoost (адаптивное обучение)
- Пример AdaBoost
- Теория градиентного бустинга
- Пример градиентного бустинга


Основная формула для Бустинга: 

$$
F_T(x) = \sum_{t=1}^{T} f_t(x)
$$
$$
f_t(x) = \alpha_t h(x)
$$

$f_t(x)$ - отдельная модель, weak learner

Собираем в общий, *сильный* ансамбль моделей

Т - количество слабых моделей, гиперпараметр

$H(x)$ - гипотеза отдельной модели о том, как может выглядеть результат модели. Может быть любим алгоритмом ML (по сути какой класс будет предсказан)

t - порядковый номер

Чем лучше результат, тем выше $\alpha_t$

Можно вычислить ошибку для каждого шага:

$$
E_t = \sum_{i} E[F_{t-1}(x_i) + \alpha_t h(x_i)]
$$

Может ли ансамбль из слабых (простых) моделей стать сильной моделью?

Для деревьев решений мы можем в качестве h(x) использовать простые деревья, объединив их потом с некоторыми коэффицентами

## **AdaBoost - Адаптивный бустинг**

Использует ансамбли слабых моделей и затем объединяет их с помощью взвешенной суммы.

Адаптация состоит в том, что ранее созданные слабые модели используются при построении следующих слабых моделей, уточняя их поведение для неправильно классифицированных точек.

**Слабая модель (Weak Learner)** - слишком простая модель, которая сама по себе не дает достаточно хороших результатов на данных.

В отличие от одного дерева решений, которое обучается сразу *на всех данных*, AdaBoost агрегирует несколько слабых моделей, позволяя обучать результирующуюу ансамблевую модель *постепенно*

Как мы можем объединять между собой простейшие деревья?

Деревья создаются один за другим, а не сразу

**Терминология для алгоритма**

•   Точки данных $x_1 \dots x_n$

•   Требуемые результаты $y_1 \dots y_n, y \in \{-1,1\}$

•   Начальные веса $w_{1,1} \dots w_{n,1}$ равны $\frac{1}{n}$

•   Функция ошибок $E(f(x), y, i) = e^{-y_i f(x_i)}$

•   Слабые модели $h: x \to \{-1,1\}$
