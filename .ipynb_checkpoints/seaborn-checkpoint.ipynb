### **Scatter Plots - диаграммы рассеяния**

Scatter Plots показывают связь между двумя *непрерывными* переменными. 

**Непрерывная переменная** - числовая переменная, которая может принимать любые промежуточные значения.

Примеры: возраст, высота, зарплата, температура. Возраст: 24 года; детальнее: 24 года и 2 месяца; еще детальнее: 24 года, 2 месяца и 10 дней и т.д.

***Непрерывные VS дискретные(категориальные) переменные***

```python
df.head()
```

```python
df.shape
```

HUE задает *третье измерение* двумерному графику

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, 
                s=200,alpha=0.7, style='level of education',
                hue='level of education'
               )
#plt.savefig('MyScatterPlot.png')
#plt.show() для скриптов .py
# hue='level of education', palette='Dark2'
```

- hue - название колонки, по которой график будет раскрашен разными цветами. Обычно берут категориальную переменную. Если взята непрерывная, будет создан градиент цвета
- palette - раскраска точек, много разных палитр
- size - то, на осное какой колонки будет настроен размер точек
- s - общий размер точек
- alpha - прозрачность точек, от [0,1]
- style - настроит отдельный маркер для каждого вида указанной колонки. Только категориальная колонка, ведь разные маркеры не связаны друг с другом
- color/c - изменить цвет всех точек

### **Distribuiton Plots - графики распределения вероятностей**

**Distribuiton Plots** отображают одну непрерывную переменную и помогают визуализировать атрибуты этой переменной (среднее значение, дисперсиию и т.д.)

- RUG Plot
- Histogram
- KDE Plot (Kernel Density Estimation)

**Rug Plot - простейший из графиков распределения**

Для каждой записи мы ставим черточку на оси абсцисс, ось ординат не используется. 

Легко просматриваются крайние значения, outliers, т.к. для них мы видим отдельные черточки.

Не видно, где много, а где мало черточек.

Если посчитаем черточки, то получим гистограмму

Если по оси у нанести количество записей в каждом из интервалов (bins), то получим гистограмму!

Значения по оси у можно нормализовать (перевести в проценты) 

Поверх гистограммы можно добавить график **KDE(Kernel Density Estimation)**, который показывает, как может выглядеть распределение вероятностей для случайной переменной, если значения этой переменной получаются из приведенного на гистограмме набора данных

**KDE(Kernel Density Estimation)** - ядерная оценка плотности

Способ оценки функции распределения некоторой непрерывной случайной величины. Способ построить непрерывный график на основе дискретных данных, по которым была построена гистограмма

Ядро здесь - выбранная функция распределения вероятности, которую мы применим в каждой точке rugplot. 

Параметры:
- kernel - функция ядра
- bandwidth - ширина полосы


```python
#plt.figure(figsize=(5,8), dpi=150)
sns.rugplot(data=df,x='salary', height=0.2) # height в % от figure
```

### **Categorical Plots - графики по категориям**

Отображают статистические метрики **для каждой из категорий** переменной

Эквивалентно вызову фукнции **groupby()**

- countplot()
    - количество строк по категориям (groupby() --> count())
- barplot()
    - общая форма отображения любой метрики по категориям

**Countplot()**

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df, x='division', hue='division')
```

```python
plt.figure(figsize=(12,4))
sns.countplot(data=df,x='level of education',hue='division', palette='Set2')
```

```python
df['level of education'].value_counts()
```

```python
sns.countplot(data=df, x='level of education', hue='training level')
```

**Barplot()**
Позволяет использовать любые агрегирующие метрики по оси у.

*!Предупреждение!*

Поскольку столбцы на них выглядят как непрерывные столбцы, то смотрящий на график человек может подумать, что ось ординат является непрерывной, но это не так! 

Всегда добавляйте описание для этих графиков

Вероятно, что данные проще будет отобразить в виде обычной таблицы

```python
#ci-confidence interval, ci='sd' - standard deviation, estimator - aggregate function
plt.figure(figsize=(12,4))
sns.barplot(data=df, x='level of education',y='salary',estimator=np.mean,errorbar='sd', hue='division')
plt.legend(loc=(1.05,0.5))
```

**Распределения по категориям**

Выше были рассмотрены графики для распределения одной переменной. Что делать, если мы хотим сравнить распределения для разных переменных? 

Например, вместо общего распределения зарплат, мы хотим сравнить распределения зарплат отдельно для каждого уровня образования?

- Boxplot
- Violinplot
- Swarmplot
- Boxenplot (Letter-Value Plot)

**Boxplot - отображает распределение непрерывной переменной**

Использует квартили

Квартили разделяют все данные на 4 равных интервала по количеству точек:
- 25% точек в нижнем интервале
- 50-й квартиль (Q2) это медиана

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, y='salary',x='level of education',hue='level of education', palette='Set2')
```

**Violinplot** очень похож на Boxplot

Отображает плотность распределения по данным с помощью KDE

Его можно представить себе как график KDE, отраженный в зеркале

```python
sns.violinplot(data=df, x='reading score', y='parental level of education',bw_adjust=0.1)
#plt.legend(loc=(1.1,0.5))
```

```python
sns.boxenplot(data=df, x='math score',y='test preparation course',hue='gender')
```

### **Comparison Plots - графики сравнения**

Двухмерные версии тех графиков, который мы изучали ранее
- jointplot()
- pairplot()

**Jointplot()**

Можем взять гистограммы для каждой переменной из scatterplot(), чтобы уточнить распределение данных по каждой переменной

Кроме scatterplot можно использовать hex/hexagon plot или двумерный KDE plot

**Pairplot()**

Позволяет сравнивать все числовые переменные в наборе данных

Автоматически создает гистограмму для каждой переменной и сравнительные графики scatterplot для всех комбинаций переменных
