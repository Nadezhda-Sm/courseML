# **Кластеризация**

**Кластеризация** — это метод обучения без учителя в машинном обучении и анализе данных, который автоматически разделяет набор объектов на группы (кластеры) на основе их схожести. Объекты внутри одного кластера максимально похожи, а объекты из разных — максимально отличаются. 

Принимает на вход **неразмеченные данные** и ищет в них похожие данные и объединяет их в группы.

Мы не знаем истинные значения целевой переменной, поэтому не можем сказать, какая кластеризация правильная, а какая -нет

Почти все методы кластеризации объединяют вместе те точки, которые находятся близко на графике судя по какой-либо метрике

Что делать в случае неочевидных или многомерных (больше 3 измерений) данных?

Кластеризация не дает "смысловых" навзаний для кластеров! Т.е. она не может скзаать, что этот класс - "собаки", этот - "кошки" и т.д.

**Основные идеи кластеризации:**

1) На основе признаков определяем, какие точки находятся близко друг к другу. Метрика расстояния мпежду точками в пространстве признаков
2) Нет известных истинных значений целевой переменной, чтобы сравнить с ними результаты кластеризации
3) *Кластеризация* - Процесс обучения без учителя, который пытается обнаружить потенциальные значения целевой переменной

**Смена парадигмы для обучения без учителя** 

- Что делать с новыми неразмеченными данными?
- Как мы поместим их в новые кластеры?
- Насколько правильно мы назначили кластеры для новых данных?

**Как мы поместим новую точку в кластер?**
- Зависит от используемого метода обучения без учителя
- на основе признаков выбирается наиболее подходящий кластер
- У нас нет способа понять, насколько правильно это сделано



- Смена парадигмы для обучения без учителя:
  - Когда мы получим кластеры, можно ли взять их как значения **у** для обучения с учителем?
    - Да! Можно сначала провести обучение без учителя, и полученные кластеры трактовать как целевую переменную для обучения с учителем (для новых точек).

- Смена парадигмы для обучения без учителя:
  - Когда мы получим кластеры, можно ли взять их как значения **у** для обучения с учителем?
    - Алгоритм не скажет нам, что означают эти кластеры. И мы не сможем понять, насколько эти кластеры значимые.

## **Кластеризация К-средних (K-Means Clustering)**

В результате должно получиться следующее:
  - Каждая точка должна принадлежать к какому-то кластеру.
  - Каждая точка может принадлежать только к одному кластеру (одна точка не может принадлежать к нескольким кластерам).

Шаг 0: Начинаем с неразмеченных данных

Шаг 1: Выбираем количество кластеров (К), которые мы будем создавать - выбрать надо до начала кластеризации

Шаг 2: случайно выбираем К различных точек. Начальные точки - центры наших классов

Шаг 3: Для остальных точек берем "цвет" ближайшей "кластерной" точки

Шаг 4: Вычисляем центры кластеров (средние значения векторов) (можно отметить эти центры на графике, но они НЕ станут новыми точками)

Шаг 5: Каждой точек назначаем кластер с ближайшим к ней центром (нек-ые точки могут поменять кластер)

Повторяем Шаги 4 и 5 до тех пор, пока все точки не перестанут менять свои цвета

Если точки не меняют свои цвета, то процесс кластеризации завершен!

- Как выбрать разумное значение К для количества классов?
- Можно ли как-то оценить, насколько то или иное значение К хорошее?

## Как выбрать К?

1) Метод локтя

сумма квадратов расстояний от точек до центров кластеров. похоже на rmse. не обязаны ее минимизировать. минимум ssd = 0 (1 точка в каждом кластере)

Sum of Squared Distances


- Мы вычисляем метрику SSD для различных K.  
- Ищем те моменты, где метрика SSD уменьшается не так сильно, как в других точках.  
- Это означает, что добавление ещё одного кластера не делает кластеры заметно более компактными — настолько, чтобы увеличивать K на единицу.

2) Метод силуэтов


Метод силуэтов:  
  ${Silhouette\_score} = \frac{b - a}{\max(a, b)}$
  

-\(a\) — среднее расстояние внутри кластера  
- \(b\) — расстояние до ближайшего соседнего кластера

Вот текст с изображения:

- Среднее для одного кластера  
- Среднее для всех кластеров  
- Значения от –1 до +1:


- 1 – отличные кластеры  
- 0 – плохие кластеры  
- < 0 – пересекающеся кластеры

Ищем точку колена на графике

```python
ssd = []

for k in range(2,10):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)

    ssd.append(model.inertia_) #Сумма квадратов расстояний от точек до центров кластеров
    
    
```

```python
ssd
```

```python
plt.plot(range(2,10), ssd, 'o--')
```
